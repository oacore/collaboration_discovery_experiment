```python
%matplotlib inline

import pickle 
import re
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def get_all_emails_with_time_of_repository(repoId):
    all_emails = []
    for page_pickle in range(1,10):
        try:
            articles = pickle.load(open('data/all_articles_on'+str(repoId)+'_'+str(page_pickle)+'.pkl','rb'))
            for page in range(0,len(articles)):
                for item in articles[page]['data']:
                    text = item['fullText']
                    article_id=item['id']
                    if 'year' in item:
                        year=item['year']
                    else:
                        continue
                    match = re.findall(r'[\w\.-]+@[\w\.-]+', text)
#                     print(match)
                    all_emails.append([article_id, year,match])
        except (OSError,IOError) as e:
            print(e)
    return all_emails

def get_all_emails_of_full_download(repoId):
    '''
    same as get_all_emails_with_time_of_repository but we use the full download pickle instead
    '''
    all_emails = []
    try:
        articles = pickle.load(open('./data/repo_'+str(repoId)+'_full_download.pkl','rb'))
        for page in range(0,len(articles)):
            for item in articles[page]['data']:
                text = item['fullText']
                article_id=item['id']
                if 'year' in item:
                    year=item['year']
                else:
                    continue
                match = re.findall(r'[\w\.-]+@[\w\.-]+', text)
#                     print(match)
                all_emails.append([article_id, year,match])
    except (OSError,IOError) as e:
        print(e)
    return all_emails

def flatten_list(unevenList):
    flat_list=[]
    for sublist in unevenList:
        aid=sublist[0]
        y=sublist[1]
        emails = sublist[2]
        for item in emails:
            flat_list.append([aid,y,item])
    return flat_list
def extract_domain(email):
    return email.split("@")[1].rstrip(",").rstrip(".")
```

```python
aberdeen_emails_with_years= get_all_emails_with_time_of_repository(1)
```

```python
aberdeen_emails_with_years_flatten=flatten_list(aberdeen_emails_with_years)
```

```python

aberdeen_df = pd.DataFrame.from_records(aberdeen_emails_with_years_flatten)
```

```python
aberdeen_df['domain']=aberdeen_df.apply(lambda row: extract_domain(row[2]),axis=1)
```

```python
aberdeen_df.columns = ["articleId","year","author_email","author_domain"]
aberdeen_df['external']=aberdeen_df['author_domain']!='abdn.ac.uk'
```

```python
aberdeen_internal_df = aberdeen_df[aberdeen_df.external==False].groupby("year").size().rename('counts').reset_index(name='counts')
aberdeen_external_df = aberdeen_df[aberdeen_df.external==True].groupby("year").size().rename('counts').reset_index(name='counts')
aberdeen_internal_df = aberdeen_internal_df[aberdeen_internal_df.year>=2004]
aberdeen_external_df = aberdeen_external_df[aberdeen_external_df.year>=2004]
```

```python
ind = np.arange(len(aberdeen_internal_df.year))
width = 0.27       # the width of the bars

fig, ax = plt.subplots()

internal = aberdeen_internal_df['counts'].tolist()
external = aberdeen_external_df['counts'].tolist()

rects1 = ax.bar(ind, internal, width, color='r')
rects2 = ax.bar(ind+width, external, width, color='g')

ax.set_ylabel('Count of publications')
ax.set_title('Internal/external collaborations over time')
ax.set_xticks(ind+width)
ax.set_xticklabels( aberdeen_internal_df.year.tolist() )
ax.legend( (rects1[0], rects2[0]), ('internal', 'external') )

fig.set_size_inches(18.5, 10.5)

plt.show()
```

```python
import os 
def plot_collaboration_over_time_of_repo(repoId,domain):
    emails_with_years=[]
    if os.path.exists('./data/repo_'+str(repoId)+'_collaboration_frame.pkl'):
        df = pickle.load(open('./data/repo_'+str(repoId)+'_collaboration_frame.pkl','rb'))
    else:
        emails_with_years = get_all_emails_of_full_download(repoId)
        emails_with_years_flatten=flatten_list(emails_with_years)
        df = pd.DataFrame.from_records(emails_with_years_flatten)
        df['domain']=df.apply(lambda row: extract_domain(row[2]),axis=1)
        df.columns = ["articleId","year","author_email","author_domain"]
        df['external']=df['author_domain']!=domain
        pickle.dump(df,open('./data/repo_'+str(repoId)+'_collaboration_frame.pkl','wb'),pickle.HIGHEST_PROTOCOL)
        
    internal_df = df[df.external==False].groupby("year").size().rename('counts').reset_index(name='counts')
    external_df = df[df.external==True].groupby("year").size().rename('counts').reset_index(name='counts')
    internal_df = internal_df[internal_df.year>=2004]
    external_df = external_df[external_df.year>=2004]
    ind = np.arange(len(internal_df.year))
    width = 0.27       # the width of the bars

    fig, ax = plt.subplots()

    internal = internal_df['counts'].tolist()
    external = external_df['counts'].tolist()

    rects1 = ax.bar(ind, internal, width, color='r')
    rects2 = ax.bar(ind+width, external, width, color='g')

    ax.set_ylabel('Count of publications')
    ax.set_title('Internal/external collaborations over time')
    ax.set_xticks(ind+width)
    ax.set_xticklabels( internal_df.year.tolist() )
    ax.legend( (rects1[0], rects2[0]), ('internal', 'external') )

    fig.set_size_inches(18.5, 10.5)

    plt.show()
```

```python
plot_collaboration_over_time_of_repo(1,'abdn.ac.uk')
```
